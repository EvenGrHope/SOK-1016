#### Introduksjon til BNP og konjunkturer

Den økonomiske situasjonen i et land er en viktig faktor for velferden i landet.
<br> Arbeidsplasser og sysselsetting er vesentlig for befolkningens velferd og for produksjon og vekst.
<br> $Mennesker + Ressurser \rightarrow$ landets samlede produksjon (brutto nasjonalprodukt(**BNP**)).
<br>
<br> **BNP:** *Summen av alle varer og tjenester som produseres i et land i løpet av en periode, ofte et år, fratrukket de varene og tjenestene som benyttes i denne produksjonen*
<br>
<br>**BNP i Norge**
<br>BNP i 2017: 3.299 miliarder kr.
<br>Befolkning i 2017: 5.258.000
<br>BNP per person/kapita: 626.200 $\leftarrow \frac{3.299.000.000.000}{5.258.000} = \frac{BNP}{Befolkning}$
<br>
<br>**BNP per person/kapita** viser gjennomsnittelig verdiskapning per innbygger.
<br>*Tallet sier imidlertid ikke noe om hvordan verdier mellom personer i befolkninger er fordelt.*
<br>
<br> **Trend-BNP/Normalt BNP**
<br> For å kunne analysere utviklingen i BNP trenger man en referanse for hva som kan forventes å være et normalt produksjonsnivå/aktivitetsnivå.
<br> Trend-BNP/Normalt BNP er det nivået på BNP som fremkommer dersom ressursene våre; arbeidskraft, teknologi, kapital og naturressurser, anvendes på et langsiktig bærekraftig nivå. Med langsiktig bærekraftig nivå menes det at ressursene hverken overforbrukes eller at det er vesentlig ubrukt kapasitet.
<br> Vi skal nå se på hvordan BNP kan se ut over en 10-års periode.

```python
# Lager lister som skal brukes til å tegne figuren.
year       = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
gdp        = [1000,1020,1040,1030,1040,1050,1070,1060,1070,1100]
ngdp       = [1000,1010,1020,1030,1040,1050,1060,1070,1080,1090]
hgdp_year  = [2010,2011,2012]
hgdp_value = [1040,1040,1040] 
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, ax = plt.subplots()

# Gir aksene navn.
ax.set_ylabel("BNP")
ax.set_xlabel("Årstall")

# Plotter budsjettlinjen og figurativ framstilling forutsatt at konsumenten kjøper 6 av vare x_1.
ax.plot  (year, gdp,             color = "blue",   label = "Reelt BNP/Faktisk BNP")
ax.plot  (year, ngdp,            color = "green",  label = "Normalt BNP/Trend BNP", linestyle = "--")
ax.plot  (hgdp_year, hgdp_value, color = "orange", label = "Høykonjunktur")
ax.vlines(2017, 990, 1060,       color = "red",    label = "Lavkonjunktur")

# Begrenser aksene. 
ax.set(xlim = (2010,2019))
ax.set(ylim = (990,1110))

# Gir figuren en tittel.
ax.set_title("BNP")

# Gir figuren en legend og grid.
ax.legend(loc = "upper left")
ax.grid()

# Viser figuren.
plt.show()
```


**Konjunktur**
<br> Svingningen på reelt BNP/faktisk BNP kaller vi for **konjunkturer**.
<br> **Konjunkturer** er et begrep som brukes til å forklare ett lands økonomiske situasjon.
<br>
<br> **Høy- og lavkonjunktur**
<br> Høy- og lavkonjunktur sier noe om hvilken tilstand økonomien er i sett i forhold til noramlt/trend BNP.
<br> Dersom reelt BNP ligger over normalt BNP befinner landet seg i det vi kaller **høykonjunktur**.
<br> **Høykonjunktur:** Landets aktivitetsnivå er større enn hva som er langsiktig opprettholdbart. Kjennetegn på dette er at produksjonen er høy og arbeidsledigheten lav.
<br> Dersom reelt BNP ligger under normalt BNP befinner landet seg i det vi kaller **lavkonjunktur**.
<br> **Lavkonjunktur:** Landets aktivitetsnivå er lavere enn hva som er langsiktig opprettholdbart. Kjennetegn på dette er at produksjonen er lav og arbeidsledigheten høy.
<br>
<br> **Opp- og nedgangskonjunktur**
<br> Opp- og nedgangskonjunktur sier noe om når konjunkturene snur. 
<br> **Oppgangskonjungtur:** Lokalt minimumspunkt $\rightarrow$ lokalt maksimumspunkt. Dette kan vi se på figuren mellom f.eks. årene 2010 (lokalt min. punkt) til 2012 (lokalt maks. punkt).
<br> **Nedgangskonjungtur:** Lokalt maksimumspunkt $\rightarrow$ lokalt minimumspunkt. Dette kan vi se på figuren mellom f.eks. årene 2016 (lokalt maks. punkt) til 2017 (lokalt min. punkt).

#### Nominell/reelt BNP og BNP-deflatoren

<br> **Nominell og reelt BNP**
<br> I nasjonalregnskapet måles størrelsene (BNP, NNP, privat konsum, offentlig konsum, imports, eksport) i nominelle/løpende priser (prisen på aktuelle tidspunkt).
<br> **Nominell BNP:** BNP målt i det aktuelle årets markedspriser.
<br> Hvordan endrer volum (mengder) av varer og tjnester seg over tid?
<br> *Økning i nominelt BNP kan både skyldes økt produksjon, men også økte/reduserte priser.*
<br> **Reelt BNP (BNP-volum):** BNP målt i faste priser (bruker priser fra basisåret til å regne ut BNP for andre år). Reelt BNP viser hvor mye av endringene i BNP som skyldes volumet varer og tjenester som er produsert.
<br>
<br> **Eksempel:**
<br> Se for deg at Norge kun produserer laks.
<br> BNP er da lik verdien av all laks som blir produsert. 
<br> **2010:**
<br> **Antall kg:** 10 kg
<br> **Pris pr. kg:** 100 kr
<br>
<br> **2011:** 
<br> **Antall kg:** 12 kg
<br> **Pris pr. kg:** 110 kr
<br>
<br> **Nominell BNP - BNP målt i det aktuelle årets markedspris**
<br> Nominell BNP i 2010: $10 * 100 = 1000 kr$
<br> Vi kan se at produksjon av laks gikk opp fra 10 kg i 2010 til 12 kg i 2011, for å finne volumveksten kan vi bruke denne formelen:
<br> **Prosentvis endring** $ = \frac{verdi_(ny)-verdi_(gammel)}{verdi_(gammel)}* 100%$
<br> **20 %** $ = \frac{12-10}{10)}* 100%$
<br> Vi kan også se at prisen pr. kg. gikk opp fra 100 kr i 2010 til 110 kr i 2011, vi kan bruke samme formel for å finne prisendringen på laks:
<br> **10 %** $ = \frac{110-100}{100)}* 100%$
<br> Nominell BNP i 2011: $12 * 110 = 1320 kr$
<br> For å finne den prosentvise endringen i nominell BNP kan vi bruke samme formel som tidligere:
<br> **32 %** $ = \frac{1320-1000}{1000)}* 100%$
<br>
<br> **Reelt BNP - BNP målt i faste priser**
<br> Vi bruker 2010 prisene (100 kr) på laks som basisår.
<br> For å finne reelt BNP kan vi bruke denne formelen:
<br> Reelt BNP det aktuelle året $= Volum(år) * Pris(basisår)$
<br> Reelt BNP (2011) $= 12 kg * 100 = 1200 kr$
<br> Reelt BNP (2010) = Nominelt BNP i basisåret = 1000 kr
<br> Vi kan da finne den prosentvise endringen i reelt BNP fra 2010 til 2011 med bruk av samme formel som isted:
<br> **20 %** $ = \frac{1200-1000}{1000)}* 100%$
<br>
<br> **BNP-deflatoren - Mål på prisvekst i BNP**
<br> Formelen for BNP-deflatoren er som følger:
<br> BNP-deflatoren $= \frac{nominellBNP}{reeltBNP}$
<br> Visst BNP-deflatoren > 1 = Prisene har økt
<br> Visst BNP-deflatoren < 1 = Prisene har sunket
<br> Visst vi bruker BNP-deflatoren på 2011 eksempelet over får vi følgende:
<br> $\frac{1320}{1200} = 1.1$ 
<br> $1.1 > 1 =$ Prisene har økt.
<br> Visst vi bruker BNP-deflatoren på 2010 eksempelet over får vi følgende:
<br> BNP-deflatoren i basisåret = 1
<br> Vi kan da se at fra 2010 til 2011 økte prisene med 10%

#### Nasjoners velstand og økonomisk vekst

Ofte når vi skal se på økonomisk vekst bruker vi BNP per innbygger/kapita. 
|Land|1913|1950|1997|2014|

<br> På tabellen over har vi brukt USA som referanse, det vil si at de andre landene er målt i forhold til BNP per innbygger i USA.
<br> BNP per innbygger er som nevnt tidligere et mål på matriell velstand.
<br>
<br> **Institusjoner**
<br> Institusjoner er ikke bare bygg, organisasjoner og stiftelser.
<br> Når økonomer snakker om gode økonomiske institusjoner mener de veletablerte markeder, lover, regler og reguleringer som påvirker adferden til deltakerne i økonomien (konsumenter, banker, bedrifter). 
<br> I samfunnsøkonomien brukes ordet mer abstrakt og omfatter også institusjoner, detter er normer i det samfunnet. 
<br> Eksempler på institusjoner er: skatte- og trygdesystemet, lover tilknyttet arbeidslivet, åpne og konkuransedyktige markeder.
<br> Vi har også politiske instutisjoner som lover, regler og praksis når det kommer til politisk deltakelse, ytrringsfrihet og rettssikkerhet.
<br> Instutisjoner er kort forklart "spilleregler" i et samfunn.
<br>
<br> Forskning har visst at fattige land har lavere kvalitet på økonomiske og politiske institusjoner enn rikere land.
<br> Eksempler på lavere kvalitet er: mangel på lover og regler, korrupsjon, krig, politiske konflikter, mangel på velfungerende demokrati.
<br>
<br> **Konvergens og divergens**
<br> To sentrale begreper når vi snakker om den økonomiske utviklingen til land er konvergens og divergens.
<br> **Konvergens:** Et land har tatt igjen et annet eller flere land.
<br> Visst vi ser på tabellen over igjen kan vi se at Norge i 1913 hadde bare 46% av BNP per innbygger i forhold til USA. Norge har fram til 1997 tatt igjen det forsprange og vare på 86% av BNP per innbygger i forhold til USA. Det vil si at Norge har konvergert mot USA i perioden 1913-1997.
<br> **Divergens:** Det blir større forskjell mellom land. 
<br>På tabellen over kan vi se at Kina har divergert i forhold til USA i perioden 1913-1997 der de gikk fra 10% til 9% av BNP per person.

#### Oljefondet

På tabellen over kan vi se at Norge har hatt en solid økonomisk vekst, dette er mye takke være petroliums virksomheten. 
<br> Vi skal nå se litt på oljefondet/statens pensjonsfond utland.
<br> Oljefondet har to hovedformål:
<br> **1) Sikre god og langsiktig forvaltning av oljeinntektene.**
<br> Det første formålet handler om at vi vet at inntektene fra olje- og gassressursene varer ikke evig. $\rightarrow$ Formuen skal kommer framtidige genrasjoner til gode.
<br> **2) Sikre god og stabil utvikling i norsk økonomi.**
<br> Handlingsregelen: Regel for hvor mye penger som kan tas ut av Oljefondet og brukes i norsk økonomi.
<br> Alle inntektene fra oljen går inn i oljefondet, mens bruken avhenger av den forventede avkastningen på fondet. Den forventende avkastningen var tidligere på 4% men ble endret i 2017 til 3%. Så handlingsregelen handler om at i dårlige tider kan vi bruke litt mer en 3%, mens i gode tider litt mindre en 3%. På denne måten er oljefondt med på å stabilisere norsk økonomi. 
<br> Norge har vært heldige med oljeressursen, men også dyktige til å ta hånd om den. Dette kan skyldes gode økonomiske institusjoner.
